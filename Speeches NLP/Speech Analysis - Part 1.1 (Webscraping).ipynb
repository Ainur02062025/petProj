```python
# import libraries
import pandas as pd
import requests
import bs4
import re
```

```python
url = 'https://thevideoink.com/speeches-by-sir-david-attenborough-on-climate-change/'
```

```python
print(url)
```

```python
# make test resquest
test_res = requests.get(url)
test_res
```

```python
# Review the text for identifier tag
soup = bs4.BeautifulSoup(test_res.text,'lxml')
soup
```

Script body is found in '<'div class="left relative", "id":"content-main"'>'

```python
script = soup.find('div', {"class":"left relative", "id":"content-main"})

script
```

```python
Lst = []
for p in script.find_all('p'):
    script_dict = {}
    script_dict['text'] = p.text
    Lst.append(script_dict)
df = pd.DataFrame(Lst)
df
```

```python
# review
df['text'][0]
```

```python
# trim and get only the speakers transcript we need
sub_df = df.drop(index=[0])
```

```python
# reset index
sub_df.reset_index(drop=True, inplace=True)
```

```python
# Index 0 to 8 was a short speech made in 2018, at Poland.
# Index 9 onwards was made in Davos.
sub_df['text']
```

```python
# Replace ?/! with .
specialchars = '[?!]'
# create an interim column of speech after replacement
sub_df['cleaned_text'] = sub_df['text'].apply(lambda x: re.sub(specialchars,'.',x))
```

```python
full_text1 = ' '.join(sub_df['cleaned_text'][:9])
full_text2 = ' '.join(sub_df['cleaned_text'][9:])
```

```python
# Write to text file
with open("Poland2018.txt", mode = "w") as f:
    f.write(full_text1)
with open("Davos.txt", mode = "w") as f:
    f.write(full_text2)
```
